```python
rioOlympics = pd.read_csv("utility/data/athletes.csv")
rioOlympics.columns
```

```python
#Bronze medal count for females = 321*1+2*5 = 331 
#Bronze medal count for males = 361*1+2*6 = 373 

groupRioBronze = rioOlympics.groupby(['sex','bronze'])

groupRioBronze = groupRioBronze.size()

groupRioBronze.reset_index(name='Observations')
```

```python
#Silver medal count for females = 321*1+2*5 = 317
#Silver medal count for males = 321*1+2*5 = 338

groupRioSilver = rioOlympics.groupby(['sex','silver'])

groupRioSilver = groupRioSilver.size()

groupRioSilver.reset_index(name='Observations')
```

```python
#Gold medal count for females = 291*1+2*13 = 318
#Gold medal count for males = 321*1+2*5 = 348
groupRioGold = rioOlympics.groupby(['sex','gold'])

groupRioGold = groupRioGold.size()

groupRioGold.reset_index(name='Observations')
```

```python
features_to_exclude = ['id', 'height', 'weight']
rioDuplicates = rioOlympics.drop(columns=features_to_exclude)
rioOlympicsClean= rioDuplicates.drop_duplicates()

medalByCountryRio = rioOlympicsClean.groupby(['nationality'])
medalByCountryRio = medalByCountryRio.sum()
medalByCountryRio
```

```python
athleteByCountryRio = rioOlympics.groupby(['nationality'])
athleteByCountryRio= athleteByCountryRio.size()
athleteByCountryRio.reset_index(name='AthleteCount')
# athleteByCountryRio['A']
```

```python
athleteByCountryRio.to_csv('utility/data/2016-athleteCountByCountry.csv')
```

```python
groupRioGender = rioOlympics.groupby(['sex'])
groupRioGender = groupRioGender.size()
group = groupRioGender.reset_index(name='AthleteCount').T
femaleCount = group[0].AthleteCount
maleCount = group[1].AthleteCount
totalCount = femaleCount + maleCount
```

```python
gender = pd.DataFrame({'female': [femaleCount],
                   'male': [maleCount],
                   'total': [totalCount]})
```

```python
gender.to_csv('utility/data/2016-gender.csv')
```
